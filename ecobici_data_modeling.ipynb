# data modeling

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
viajes = pd.read_csv('viajes_clean.csv',header=None)
viajes.columns = ['dia','hora','usos']
```

```python
viajes.head()
```

```python
X = viajes[['dia','hora']]
y = viajes['usos']
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
```

```python
# Train the model using the training sets
regr.fit(X, y)

# Make predictions using the testing set
y_pred = regr.predict(X)
```

```python
viajes['pred'] = y_pred
viajes.head()
```

```python
viajes.usos.plot()
viajes.pred.plot()
plt.show()
```

```python
from sklearn import neighbors

n_neighbors = 5
knn = neighbors.KNeighborsRegressor(n_neighbors, weights='distance')
knn.fit(X,y)
pred_knn = knn.predict(X)
viajes['pred_knn'] = pred_knn
viajes.usos.plot()
viajes.pred_knn.plot()
plt.show()
```

```python
# X.head()
import numpy as np
x_new = np.array([[2,12]])
knn.predict(x_new)
```

```python
from sklearn import svm

svm_mod = svm.SVR()
svm_mod.fit(X,y)
pred_svm = svm_mod.predict(X)
viajes['pred_svm'] = pred_svm

viajes.usos.plot()
viajes.pred_svm.plot()
plt.show()
```

```python
from sklearn import tree

mod_tree = tree.DecisionTreeRegressor()
mod_tree = mod_tree.fit(X, y)

pred_tree = mod_tree.predict(X)
viajes['pred_tree'] = pred_tree
viajes.usos.plot()
viajes.pred_tree.plot()
plt.show()
```
